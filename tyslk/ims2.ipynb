```python
import fitz  # PyMuPDF
import os

def save_images_from_pdf(pdf_path, output_dir):
    # Open the PDF document
    document = fitz.open(pdf_path)
    
    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)
    
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        images = page.get_images(full=True)
        
        if images:
            for img_index, img in enumerate(images, start=1):
                xref = img[0]
                base_image = document.extract_image(xref)
                image_bytes = base_image["image"]
                image_ext = base_image['ext']
                
                # Save the image
                img_path = os.path.join(output_dir, f"page_{page_num + 1}_img_{img_index}.{image_ext}")
                with open(img_path, 'wb') as f:
                    f.write(image_bytes)
                
                print(f"Image {img_index} saved to {img_path}")
    
    document.close()

# Usage
pdf_path = 'C:\\Users\\ravichandrav\\Downloads\\monopoly.pdf'
output_dir = 'C:\\Users\\ravichandrav\\Desktop\\Data\\extracted_images1'
save_images_from_pdf(pdf_path, output_dir)
```
