```python
import os
import fitz  # PyMuPDF

def extract_images_from_pdf(pdf_path, output_folder):
    """
    Extracts images from a PDF using PyMuPDF and saves them locally.

    Args:
        pdf_path: Path to the PDF file.
        output_folder: Path to the folder where images will be saved.
    """
    doc = fitz.open(pdf_path)

    for page_number in range(len(doc)):
        for image_index, img in enumerate(doc.get_page_images(page_number)):
            xref = img[0]
            # Extract image data
            image_data = doc.extract_image(xref)
            image_bytes = image_data["image"]
            image_ext = image_data["ext"]  # Extract image extension (e.g., 'png', 'jpg')

            # Create filename with page number and index
            filename = f"{os.path.splitext(os.path.basename(pdf_path))[0]}_page_{page_number+1}_image_{image_index+1}.{image_ext}"
            print(filename)
            output_path = os.path.join(output_folder, filename)

            # Save image
            with open(output_path, "wb") as f:
                f.write(image_bytes)

            # Print filename
            print(f"Saved image: {filename}")

    print(f"Extracted images from {pdf_path} and saved to {output_folder}")

# Example usage
pdf_path = r"C:\\Users\\ravichandrav\\Desktop\\DW251 etc._385930-03.pdf"
output_folder = r"D:\\extracted_images"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

extract_images_from_pdf(pdf_path, output_folder)
```
